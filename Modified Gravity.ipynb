# Modified Gravity in Local Group Dwarf Galaxies

We would like to explore the idea of using Local Group dwarf galaxies to probe theories of modified gravity. As a starting point, we are interested in extending the study of $f(R)$ gravity models with chameleon screening, such as those studied by [Jain et al. 2012](http://arxiv.org/pdf/1204.6044v2.pdf), [Jain and VanderPlas 2011](http://arxiv.org/pdf/1106.0065v1.pdf), and [Vikram et al. 2013](http://arxiv.org/abs/1303.0295). At the most basic level, our goal would be to see if we can use Local Group dwarf galaxies to extended the sensitivity region in the [Snowmass Cosmic Frontier Report](http://arxiv.org/abs/1401.6085) to lower field values and smaller length scales.

Following Section 1 in [Jain and VanderPlas (2011)](http://arxiv.org/pdf/1106.0065v1.pdf) we choose to work in a modified gravity scenario with chameleon screening (we would like to extend this to other screening mechanisms as well). 
Previous studies have focused on dwarf spiral galaxies with rotational velocity of ~50 - 100 km/s and significant HI gas content.
Local Group dwarf spheroidal galaxies are a factor of 5 - 10 less massive (velocity dispersion of 3 - 15 km/s) and could be used to probe smaller values of $f_{R0}$.
The Newtonian potential at the virial radius of a dwarf galaxy scales roughly as $|\Phi_{N}| \propto M/r_{\rm vir} \propto M^{2/3} \propto V_{circ}^2$. 
The Milky Way has $|\Phi_{N}| \approx 10^{-6}$ (with c=1) and a circular velocity of $V_{circ} \sim 200$ km/s. The dwarf galaxies study previously have $V_{circ} \sim 50$ km/s and $|\Phi_{N}| \sim 10^{-7}$. 
Local Group dwarf galaxies have a velocity dispersion of $\sigma_v \sim 3-15$ km/s and would have $|\Phi_{N}| \sim 10^{-8}$. The background field value of modified gravity has to be $f_{R0} \gtrsim 10^{-8}$ to explain dark energy, so this is a natural stopping point.

There are some important physical and observational differences between the Local Group dwarfs and many of the objects studied previously. 
* The LG dwarfs are pressure supported spheroidal galaxies lacking stellar disks or measured rotations
* Dwarf galaxies within the MW virial radius have no appreciable HI gas content (e.g. [Grcevich & Putnam 2009](http://arxiv.org/abs/0901.4975)). Dwarfs outside the MW virial radius do contain gas, but it may not have a measured rotational velocity (need to look into this further).
* A large number of dwarf galaxies in the Local Group can be resolved into their individual stars. This allows for fits to the spatial morphology of various stellar populations.
* Velocity dispersions to be calculated from multi-object spectroscopy of individual stars. The velocities of individual stars can be measured with an accuracy of several km/s (1 - 3 km/s) and velocity dispersions can be measured to an accuracy of 1.5 km/s (20 stars) to better than 0.5 km/s (100+ stars).
* The mass distribution, and thus the local estimate of the $\Phi_{N}$, should be more accurately known within the LG relative to larger distances.

## Estimating Velocity Dispersions

If RGB stars are unscreened, they would feel a 5th force that would modify Newton's constant. From Eqn 1 & 2 of [Jain et al. (2012)](http://arxiv.org/abs/1204.6044), the modification would be $G \rightarrow G(1+\alpha_c)$ where $\alpha_c$ sets the strength of the 5th force in unscreened regions. For partially screened objects $G(r) = G [ 1 + \alpha_c (1 - M(r_s)/M(r))]$ where $M(r)$ is the mass enclosed. The value of $\alpha_c$ can range from ?? to ??, but conventional values explored in the literature are $\alpha_c = 1/3$ and $\alpha_c = 1$ ([Jain et al. (2012)](http://arxiv.org/abs/1204.6044)).

In this case that RGB stars are completely unscreened (an over-optimistic situation, since the cores of these stars are still screened) they would feel an added force as $G \rightarrow 4/3 G$. This can be interpreted as a increase in the "effective mass" of the dwarf. The measured velocity dispersion, $\sigma_v \propto \sqrt{M}$, will see a proportional increase for the RGB sample of $\sigma_{vRGB} = \sqrt{4/3} \sigma_{vMS}$ or a 15% increase in the velocity dispersion. For a Local Group dwarf galaxy with a MS velocity dispersion of $\sigma_{vMS} = 10$ km/s, this correpsonds to a $\Delta \sigma = 1.5$ km/s. To get a sense for our observational abilities, from [Simon & Geha (2007) ](http://arxiv.org/abs/0706.0516) the measurement uncertainty on the velocity dispersion is $\delta \sigma \sim 1$ km/s for 25 stars (CVn II), $\delta \sigma \sim 0.7$ km/s for 60 stars (Com) and $\delta \sigma \sim 0.4$ km/s for 200 stars (CVn I).

## Estimating Spatial Offsets

The 5th force felt by unscreened RGB stars would also lead to a spatial offset between the distribution of RGB and MS stars. This offset would occur as RGB stars felt an increased gravitational attraction during interactions between dwarf galaxies and other massive objects. The direction of this offset will be alligned with the net force on the dwarf galaxy. Unfortunately, for many nearby galaxies this force will be aligned with the line-of-sight to the Milky Way and would be very difficult (impossible?) to measure. We would therefore be interested in the interaction between dwarf galaxies and other large Local Group bodies (i.e., M31) or the interaction between dwarfs and other nearby dwarf companions (potentially satellites of satellites).

## Estimating Potentials

Here we estimate a few potentials relavent to the study of dwarf galaxies. Eqn 2.1 in [Vikram et al. 2013](http://arxiv.org/abs/1303.0295) describes to conditions under which a galaxy will be screened:

$\frac{|\Phi_{int}|}{c^2} > \frac{3}{2} f_{R0}$ and $\frac{|\Phi_{ext}|}{c^2} > \frac{3}{2} f_{R0}$.

The first term describes self-screening from the galaxy itself and the second term is related to screening from the local (and perhaps not so local) environment. The external potential is the sum of local contributors, $\Phi_{ext} = \sum_{d_i < \lambda_C + r_i} \frac{G M_i}{d_i}$ , where $d_i$ is the distance to a neighboring galaxy with virial mass $M_i$ and radius $r_i$, and $\lambda_C$ is the Compton wavelength of the background field, $\lambda_C \approx 32 \sqrt{\frac{|f_{R0}|}{10^{-4}}}$ Mpc ([Schmidt et al. 2009](http://arxiv.org/abs/0908.2457)). The Compton wavelength essentially sets the maximum range of the fifth force, setting the "horizon" size for massive objects that can contribute to screening from the external potential. 

For $f_{R0} < 10^{-6}$, then $\lambda_C \lesssim 3.2$ Mpc. This means that when constructing our map of the local potential we would need to encorporate massive objects on roughly the scale of the Local Group (i.e., the Milky Way, Andromeda, etc.), but not larger scales such as the Virgo cluster or the Laniakea supercluster (this is a good thing...).

```python
# Newtonian potential
G = 4.302e-3 # pc * Msun^-1 * (km/s)
c = 3e5 # km/s

def PhiN(m,r):
    return G/c**2 * m/r

# For the Sun
M = 1 # Msun
R = 6.957e8/(3.086e16) # In pc
print "Potential at Solar surface: %.1e"%PhiN(M,R)

Rs = 20 # Milky Way NFW scale radius
R1 = 30 # Milky Way disk radius
R2 = 300 # Milky Way virial radius
Mvir = 1.2e12 # Milky Way virial mass (Msun)
Mstar = 1e11 # Milky Way stellar mass (Msun)

# Find the mass of the Milky Way inside the disk
def Mnfw(Rmax,Rs=20):
    return np.log((Rs+Rmax)/Rs) - Rmax/(Rs+Rmax)

Mdisk = Mvir * Mnfw(R1)/Mnfw(R2) + Mstar # Mass within MW disk (Msun)

### Note that this is not the conventional screening potential
### since it is not measured at the virial radius
M = Mdisk
R = 30e3 # Milky Way disk radius kpc
print "Potential outside Milky Way disk: %.1e"%PhiN(M,R)

M = Mvir  # Virial mass of the MW (Msun)
R = 300e3 # Virial radius of the MW (pc)
print "Potential at the Milky Way virial radius: %.1e"%PhiN(M,R)

# For Virgo Supercluster
M = 1e15 # Msun
R = 33e6 # Mpc
print "Potential from Virgo Supercluster: %.1e"%PhiN(M,R)

# For Laniakea Supercluster
M = 1e17 # Msun
R = 77e6 # Mpc
print "Potential from Laniakea Supercluster: %.1e"%PhiN(M,R)
```

Note that both the Virgo and Laniakea superclusters have local Newtonian potentials that are larger than the vlaue of $f_{R0}$ that we would like to probe. However, both are also much farther than a Compton wavelength from the Local Group (a factor of ~10x and ~25x respectively).
