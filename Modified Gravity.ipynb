# Modified Gravity in Local Group Dwarf Galaxies

We would like to explore the idea of using Local Group dwarf galaxies to probe modifications to gravity. Specifically, we are interested in 5th force $f(R)$ gravity models with chameleon screening, such as those studied by [Jain et al. 2012](http://arxiv.org/pdf/1204.6044v2.pdf), [Jain and VanderPlas 2011](http://arxiv.org/pdf/1106.0065v1.pdf), and [Vikram et al. 2013](http://arxiv.org/abs/1303.0295). Our basic goal would be to see if we can take the dwarf galaxie constraints reported in the [Snowmass Cosmic Frontier Report](http://arxiv.org/abs/1401.6085) and extend them to lower field values and smaller size scales.

Previous studies have focused on dwarf spiral galaxies with rotational velocity of ~50 km/s and significant HI gas content.  Studies of Local Group dwarf spheroidal galaxies would probe objects that are a factor of 5 - 10 less massive. The Local Group dwarf spheroidals are pressure supported with velocity dispersions of 3 - 15 km/s, and may have no measureable neutral gas content.

Following Section 1 in [Jain and VanderPlas (2011)](http://arxiv.org/pdf/1106.0065v1.pdf) we choose to work in a modified gravity scenario with chameleon screening (we would like to extend this to Vainshtein screening). The Newtonian potential at the virial radius of a dwarf galaxy scales roughly as $|\Phi_{N}| \propto M/r_{\rm vir} \propto M^{2/3} \propto v_{c}^2$. The Milky Way has $|\Phi_{N}| \approx 10^{-6}$ (with c=1) and a circular velocity of $v_c \sim 200$ km/s. The dwarf galaxies study previously have $v_c \sim 50$ km/s and $|\Phi_{N}| \sim 10^{-7}$. Local Group dwarf galaxies have $\sigma_v \sim 3-15$ km/s and would have $|\Phi_{N}| \sim 10^{-8}$. The background field value of modified gravity has to be $f_{R0} \gtrsim 10^{-8}$ to explain dark energy, so this is a natural stopping point.

One of technique applied in previous studies has been to search for offsets between the stellar and gaseous components of dwarf spiral galaxies. ## Estimating Velocity Dispersions

If RGB stars are unscreened, they would feel a 5th force that would modify Newton's constant. From Eqn 1 & 2 of [Jain et al. (2012)](http://arxiv.org/abs/1204.6044), the modification would be $G \rightarrow G(1+\alpha_c)$ where $\alpha_c$ sets the strength of the 5th force in unscreened regions. For partially screened objects $G(r) = G [ 1 + \alpha_c (1 - M(r_s)/M(r))]$ where $M(r)$ is the mass enclosed. The value of $\alpha_c$ can range from ?? to ??, but as a concrete example we choose $\alpha_c = 1/3$.

In this case that RGB stars are completely unscreened (an over-optimistic situation, since the cores of these stars are still screened) they would feel an added force as $G \rightarrow 4/3 G$. This can be interpreted as a increase in the "effective mass" of the dwarf. The measured velocity dispersion, $\sigma_v \propto \sqrt{M}$, will see a proportional increase for the RGB sample of $\sigma_{vRGB} = \sqrt{4/3} \sigma_{vMS}$ or a 15% increase in the velocity dispersion. For a Local Group dwarf galaxy with a MS velocity dispersion of $\sigma_{vMS} = 10$ km/s, this correpsonds to a $\Delta \sigma = 1.5$ km/s. ## Estimating Spatial Offsets

```python
# Newtonian potential
G = 4.302e-3 # pc * Msun^-1 * (km/s)
c = 3e5 # km/s

def Phi(m,r):
    return G/c**2 * m/r
```
